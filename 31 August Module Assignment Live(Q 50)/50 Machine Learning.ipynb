Difference between loc and iloc

```python


# **loc** (label-based indexing):
# * Selects data based on labels (row and column names).
# * Uses explicit index and column names.
# * Includes the endpoint when slicing.

# **iloc** (integer-based indexing):
# * Selects data based on numerical positions (row and column indices).
# * Uses integer indices.
# * Excludes the endpoint when slicing.
```
