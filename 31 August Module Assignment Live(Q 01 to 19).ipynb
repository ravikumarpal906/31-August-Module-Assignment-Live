**• What is break, continue and pass in Python?**

```python

# **break, continue, and pass in Python:**

# **break:** Terminates the current loop (for or while) prematurely and resumes execution at the next statement after the loop.

for i in range(10):
  if i == 5:
    break  # Exits the loop when i becomes 5
  print(i)
# Output: 0 1 2 3 4

# **continue:** Skips the rest of the current iteration of the loop and moves to the next iteration.

for i in range(10):
  if i % 2 == 0:
    continue  # Skips even numbers
  print(i)
# Output: 1 3 5 7 9

# **pass:** A null statement, a placeholder that does nothing. It's often used as a syntactic requirement when a statement is needed but you don't want to execute any code.

for i in range(10):
  if i == 5:
    pass  # Does nothing, just a placeholder
  else:
    print(i)
# Output: 0 1 2 3 4 6 7 8 9
```

**• What are global, protected and private attributes in Python?**

```python

# **Global, Protected, and Private Attributes in Python:**

# **Global Attributes:**

# - Variables declared outside of any function or class.
# - Accessible from anywhere within the module.

global_var = 10  # This is a global variable

def my_function():
  print(global_var)  # Accessing the global variable

my_function()  # Output: 10

# **Protected Attributes:**

# - Conventionally indicated by a single underscore prefix (e.g., `_protected_var`).
# - Intended to signal that the attribute should not be accessed directly from outside the class, but it's not enforced by the language.
# - Primarily a matter of convention and readability.

class MyClass:
  def __init__(self):
    self._protected_var = 20

obj = MyClass()
print(obj._protected_var)  # Accessing the protected attribute (still possible)

# **Private Attributes:**

# - Indicated by a double underscore prefix (e.g., `__private_var`).
# - Python performs name mangling to make them less accessible from outside the class.
# - Name mangling changes the attribute name to `_ClassName__private_var`, making it harder to access directly.

class MyClass:
  def __init__(self):
    self.__private_var = 30

obj = MyClass()
# print(obj.__private_var)  # This would raise an AttributeError
print(obj._MyClass__private_var)  # Accessing the private attribute (still possible, but discouraged)

# Key Points:

# - Python doesn't strictly enforce private attributes like some other languages.
# - Name mangling is a mechanism to discourage direct access, but it's not a foolproof way to achieve true privacy.
# - The use of protected and private attributes is largely about convention and indicating the intended level of encapsulation.
```

**• What are decorators in Python? Explain it with an example.Write down its use cases.**

```python

# **Decorators in Python:**

# Decorators are a powerful way to modify the behavior of functions or classes without directly changing their source code.
# They essentially wrap a function or class, adding functionality before, after, or around the original code.

# **Example:**

def my_decorator(func):  # Decorator function
  def wrapper(*args, **kwargs):  # Wrapper function
    print("Something is happening before the function is called.")
    result = func(*args, **kwargs)  # Calling the original function
    print("Something is happening after the function is called.")
    return result
  return wrapper

@my_decorator  # Applying the decorator to a function
def say_hello(name):
  print(f"Hello, {name}!")

say_hello("Alice")

# Output:
# Something is happening before the function is called.
# Hello, Alice!
# Something is happening after the function is called.

# **Explanation:**

# 1. `my_decorator` is a function that takes another function (`func`) as its argument.
# 2. Inside `my_decorator`, there's a nested function called `wrapper`.
# 3. `wrapper` takes arbitrary arguments and keyword arguments (`*args`, `**kwargs`) to handle any parameters passed to the decorated function.
# 4. `wrapper` adds code to be executed before and after calling the original function (`func`).
# 5. `my_decorator` returns the `wrapper` function.
# 6. The `@my_decorator` syntax above the `say_hello` function applies the decorator. It's equivalent to writing:
#    `say_hello = my_decorator(say_hello)`

# **Use Cases of Decorators:**

# - **Timing Function Execution:** Measure how long a function takes to run.
# - **Logging:** Log function calls, arguments, and return values.
# - **Authentication and Authorization:** Check permissions before allowing access to a function.
# - **Caching:** Store the results of expensive function calls to avoid recomputation.
# - **Input Validation:** Validate function arguments before executing the core logic.
# - **Modifying Function Behavior:** Change the way a function behaves without altering its original code.

# **Key Points:**

# - Decorators enhance code reusability and maintainability.
# - They provide a clean way to separate concerns and add functionality without cluttering the original function's code.
# - Decorators can be nested, allowing you to apply multiple decorators to a single function.
```

How is memory managed in Python

```python

# **Memory Management in Python**

# Python uses a combination of techniques for memory management, including:

# **1. Reference Counting:**

# - Each object in Python has a reference count that keeps track of how many variables or other objects are pointing to it.
# - When the reference count of an object drops to zero, the garbage collector automatically reclaims the memory.

# **2. Garbage Collection:**

# - The garbage collector is a process that periodically identifies and reclaims memory occupied by objects that are no longer accessible.
# - It uses a cycle detection algorithm to find and collect circular references (objects referencing each other, even if they are not reachable from the main program).

# **3. Memory Pool:**

# - Python uses a private heap space to store its objects and data structures.
# - This heap is not accessible to the programmer directly.
# - The Python memory manager manages this private heap.

# **4. Dynamic Memory Allocation:**

# - Python uses dynamic memory allocation to allocate memory for objects as needed during program execution.
# - When an object is no longer needed, the memory is deallocated and returned to the system.

# **Key Components:**

# - **Python Memory Manager:** The core component responsible for allocating and deallocating memory.
# - **Garbage Collector:**  Reclaims memory occupied by unreachable objects.
# - **Reference Counting:** Keeps track of how many references point to an object.

# **Benefits of Python's Memory Management:**

# - **Automatic Memory Management:**  Programmers don't need to manually allocate or deallocate memory.
# - **Reduced Memory Leaks:** Garbage collection helps prevent memory leaks caused by unreachable objects.
# - **Improved Performance:** Efficient memory management contributes to better program performance.

# **Note:** While Python handles memory management automatically, it's still important to be mindful of memory usage, especially in large applications.
```

What is lambda in Python? Why is it used

```python

# **Lambda in Python**

# Lambda functions are small, anonymous functions defined using the `lambda` keyword.
# They can have any number of arguments but only one expression.

# **Syntax:**
# `lambda arguments: expression`

# **Example:**
add = lambda x, y: x + y
print(add(5, 3))  # Output: 8

# **Why Use Lambda Functions?**

# 1. **Conciseness:** Create functions in a single line of code, making them ideal for short, simple operations.
# 2. **Anonymous Functions:** Don't need a formal name, useful when you need a function for a short time or as an argument to another function.
# 3. **Higher-Order Functions:** Often used with functions like `map`, `filter`, and `reduce` where you need to pass a function as an argument.

# **Example with `map`:**
numbers = [1, 2, 3, 4, 5]
squared_numbers = list(map(lambda x: x**2, numbers))
print(squared_numbers)  # Output: [1, 4, 9, 16, 25]

# **Key Points:**

# - Lambda functions are expressions that evaluate to functions.
# - They are limited to a single expression, making them suitable for simple operations.
# - Commonly used for short, anonymous functions or as arguments to higher-order functions.
```

Explain split() and join() functions in Python

```python

# **split() function**

# - Splits a string into a list of substrings based on a delimiter.
# - If no delimiter is specified, it splits on whitespace (spaces, tabs, newlines).

text = "Hello, world! This is a sentence."

# Splitting on spaces
words = text.split()
print(words)  # Output: ['Hello,', 'world!', 'This', 'is', 'a', 'sentence.']

# Splitting on commas
parts = text.split(", ")
print(parts)  # Output: ['Hello', 'world! This is a sentence.']

# **join() function**

# - Joins a list of strings into a single string using a specified delimiter.

words = ['Hello,', 'world!', 'This', 'is', 'a', 'sentence.']

# Joining with spaces
sentence = " ".join(words)
print(sentence)  # Output: Hello, world! This is a sentence.

# Joining with hyphens
hyphenated = "-".join(words)
print(hyphenated)  # Output: Hello,-world!-This-is-a-sentence.

# **Key Points:**

# - `split()` is used to break down a string into smaller parts.
# - `join()` is used to combine strings together.
# - Both functions are very useful for string manipulation and text processing.
```

What are iterators , iterable & generators in Python

```python

# **Iterators, Iterables, and Generators in Python**

# **Iterable:**

# - An object that can be looped over using a `for` loop.
# - Examples: lists, tuples, strings, dictionaries, files.
# - Has a method called `__iter__()` that returns an iterator.

my_list = [1, 2, 3]
for item in my_list:  # my_list is an iterable
  print(item)

# **Iterator:**

# - An object that produces the next value in a sequence when you call `next()` on it.
# - Created using the `iter()` function on an iterable.
# - Raises a `StopIteration` exception when there are no more items.

my_iter = iter(my_list)  # my_iter is an iterator
print(next(my_iter))  # Output: 1
print(next(my_iter))  # Output: 2
print(next(my_iter))  # Output: 3
# print(next(my_iter))  # This would raise a StopIteration exception

# **Generators:**

# - Special type of iterator that is defined using a function with the `yield` keyword.
# - Generates values on the fly instead of storing them all in memory at once.
# - More memory-efficient for large sequences.

def my_generator(n):
  for i in range(n):
    yield i

gen = my_generator(5)
for item in gen:
  print(item)  # Output: 0 1 2 3 4

# **Key Differences:**

# - **Iterable:** The object you can loop over.
# - **Iterator:** The object that produces the next value.
# - **Generator:** A special kind of iterator that generates values on demand.

# **Benefits of Generators:**

# - **Memory Efficiency:**  Generate values only when needed, reducing memory consumption.
# - **Lazy Evaluation:** Values are computed only when requested, improving performance.
# - **Infinite Sequences:** Can represent infinite sequences without storing all values.
```

What is the difference between xrange and range in Python

```python
# The difference between xrange and range in Python

# In Python 2, there were two functions for generating sequences of numbers: `xrange` and `range`.

# `range` returned a list containing all the numbers in the sequence.
# `xrange` returned an xrange object that generated the numbers in the sequence on demand (similar to a generator).

# Example in Python 2:

# print(range(5))  # Output: [0, 1, 2, 3, 4]
# print(xrange(5)) # Output: xrange(5)

# The key difference was memory usage. `range` created the entire list in memory, which could be inefficient for large ranges.
# `xrange` was more memory-efficient as it generated numbers only when needed.

# In Python 3, `xrange` was removed, and `range` now behaves like the old `xrange`.
# It returns a range object that generates numbers on demand.

# Example in Python 3:

print(range(5))  # Output: range(0, 5)

# So, if you're working with Python 3, there is only `range`, and it's memory-efficient.
# If you encounter `xrange` in older Python 2 code, remember that it's similar to the `range` function in Python 3.
```

```python

# **Inheritance in Python**

# Inheritance is a mechanism in object-oriented programming that allows you to create new classes (child classes) from existing ones (parent classes).
# Child classes inherit attributes and methods from their parent classes, promoting code reuse and reducing redundancy.

# **Types of Inheritance in Python**

# 1. Single Inheritance:
#    - A child class inherits from only one parent class.

class Animal:
  def __init__(self, name):
    self.name = name

  def speak(self):
    print("Generic animal sound")

class Dog(Animal):
  def speak(self):
    print("Woof!")

my_dog = Dog("Buddy")
my_dog.speak()  # Output: Woof!


# 2. Multiple Inheritance:
#    - A child class inherits from multiple parent classes.

class Flyer:
  def fly(self):
    print("Flying...")

class Swimmer:
  def swim(self):
    print("Swimming...")

class Duck(Flyer, Swimmer):
  pass

my_duck = Duck()
my_duck.fly()   # Output: Flying...
my_duck.swim()  # Output: Swimming...


# 3. Multilevel Inheritance:
#    - A child class inherits from a parent class, which in turn inherits from another parent class.

class Vehicle:
  def __init__(self, brand):
    self.brand = brand

class Car(Vehicle):
  def drive(self):
    print("Driving...")

class ElectricCar(Car):
  def charge(self):
    print("Charging...")

my_electric_car = ElectricCar("Tesla")
print(my_electric_car.brand)  # Output: Tesla
my_electric_car.drive()       # Output: Driving...
my_electric_car.charge()      # Output: Charging...


# 4. Hierarchical Inheritance:
#    - Multiple child classes inherit from a single parent class.

class Shape:
  def calculate_area(self):
    pass  # Abstract method

class Circle(Shape):
  def __init__(self, radius):
    self.radius = radius

  def calculate_area(self):
    return 3.14 * self.radius * self.radius

class Square(Shape):
  def __init__(self, side):
    self.side = side

  def calculate_area(self):
    return self.side * self.side

my_circle = Circle(5)
my_square = Square(4)

print(my_circle.calculate_area())  # Output: 78.5
print(my_square.calculate_area())  # Output: 16


# 5. Hybrid Inheritance:
#    - A combination of two or more types of inheritance.

class A:
  def method_a(self):
    print("Method A")

class B(A):
  def method_b(self):
    print("Method B")

class C(A):
  def method_c(self):
    print("Method C")

class D(B, C):
  def method_d(self):
    print("Method D")

my_d = D()
my_d.method_a()  # Output: Method A
my_d.method_b()  # Output: Method B
my_d.method_c()  # Output: Method C
my_d.method_d()  # Output: Method D

# **Key Benefits of Inheritance:**

# - Code Reusability: Avoid writing the same code multiple times.
# - Reduced Redundancy: Promotes DRY (Don't Repeat Yourself) principle.
# - Improved Code Organization: Creates a clear hierarchy of classes.
# - Easier Maintenance: Changes in parent classes automatically reflect in child classes.
# - Extensibility: Easily add new functionality by creating child classes.
```

What is encapsulation? Explain it with an example8

```python

# Encapsulation in Python

# Encapsulation is one of the fundamental principles of object-oriented programming (OOP).
# It refers to the bundling of data (attributes) and methods (functions) that operate on that data within a single unit (class).

# Encapsulation serves several important purposes:

# 1. Data Protection:
#    - Encapsulation helps protect data from unauthorized access and modification.
#    - By restricting direct access to attributes, you can ensure that data is modified only through well-defined methods.

# 2. Data Integrity:
#    - Encapsulation helps maintain data integrity by enforcing rules and constraints on how data can be accessed and modified.
#    - This prevents accidental or intentional corruption of data.

# 3. Code Organization:
#    - Encapsulation promotes code organization by grouping related data and methods within a class.
#    - This makes code more modular, readable, and maintainable.

# Example:

class BankAccount:
  def __init__(self, account_number, balance):
    self.__account_number = account_number  # Private attribute
    self.__balance = balance              # Private attribute

  def deposit(self, amount):
    if amount > 0:
      self.__balance += amount
      print(f"Deposit successful. New balance: {self.__balance}")
    else:
      print("Invalid deposit amount.")

  def withdraw(self, amount):
    if amount > 0 and amount <= self.__balance:
      self.__balance -= amount
      print(f"Withdrawal successful. New balance: {self.__balance}")
    else:
      print("Insufficient funds or invalid withdrawal amount.")

  def get_balance(self):
    return self.__balance

# Creating a bank account object
my_account = BankAccount("1234567890", 1000)

# Accessing balance using a method
print(my_account.get_balance())  # Output: 1000

# Trying to access private attributes directly (will result in an error)
# print(my_account.__balance)  # AttributeError: 'BankAccount' object has no attribute '__balance'

# Depositing and withdrawing money using methods
my_account.deposit(500)      # Output: Deposit successful. New balance: 1500
my_account.withdraw(200)     # Output: Withdrawal successful. New balance: 1300

# Key Points:

# - Encapsulation is achieved in Python using private attributes (prefixed with double underscores `__`).
# - Private attributes cannot be accessed directly from outside the class.
# - Methods (like `deposit`, `withdraw`, `get_balance`) provide controlled access to the data.
# - Encapsulation helps ensure data protection, integrity, and code organization.
```

What is polymorphism? Explain it with an example.

```python

# Polymorphism in Python

# Polymorphism is a core concept in object-oriented programming (OOP) that refers to the ability of objects of different classes to respond to the same method call in different ways.

# The word "polymorphism" comes from Greek, meaning "many forms." In programming, it means that the same method name can be used for different objects, and each object can have its own implementation of that method.

# Benefits of Polymorphism:

# - Flexibility: Write code that can work with objects of different classes without knowing their specific types.
# - Code Reusability: Reduce code duplication by using common interfaces for different objects.
# - Extensibility: Easily add new classes and have them work with existing code without modification.

# Example:

class Shape:
  def calculate_area(self):
    pass  # Abstract method (no implementation)

class Circle(Shape):
  def __init__(self, radius):
    self.radius = radius

  def calculate_area(self):
    return 3.14 * self.radius * self.radius

class Square(Shape):
  def __init__(self, side):
    self.side = side

  def calculate_area(self):
    return self.side * self.side

# Creating objects of different shapes
circle = Circle(5)
square = Square(4)

# Calling the same method on different objects
print(circle.calculate_area())  # Output: 78.5
print(square.calculate_area())  # Output: 16

# Using polymorphism with a list of shapes
shapes = [circle, square]
for shape in shapes:
  print(shape.calculate_area())  # Output: 78.5, 16

# Key Concepts:

# - Method Overriding: Child classes can provide their own implementation of methods inherited from parent classes.
# - Duck Typing: If it walks like a duck and quacks like a duck, then it must be a duck. Python focuses on the behavior of objects rather than their specific types.
# - Interfaces (Abstract Classes): Define common methods that different classes must implement, ensuring consistency and interoperability.
```

```python
# Accept marks in five subjects
subjects = ["Subject 1", "Subject 2", "Subject 3", "Subject 4", "Subject 5"]
marks = []

for subject in subjects:
  while True:
    try:
      mark = int(input(f"Enter marks for {subject}: "))
      if 0 <= mark <= 100:
        marks.append(mark)
        break
      else:
        print("Invalid marks. Marks should be between 0 and 100.")
    except ValueError:
      print("Invalid input. Please enter a number.")

# Display marks
print("\nMarks obtained:")
for i in range(len(subjects)):
  print(f"{subjects[i]}: {marks[i]}")

def sum_of_numbers(numbers):
    total = 0
    for number in numbers:
        total += number
    return total

# Calculate total marks
total_marks = sum_of_numbers(marks)
print(f"\nTotal marks: {total_marks}")

# Calculate percentage
percentage = total_marks / 5
print(f"Percentage: {percentage:.2f}%")

# Find grade
grade = ""
match percentage:
  case percentage if percentage > 85:
    grade = "A"
  case percentage if percentage >= 75:
    grade = "B"
  case percentage if percentage >= 50:
    grade = "C"
  case percentage if percentage > 30:
    grade = "D"
  case _:
    grade = "Reappear"

print(f"Grade: {grade}")
```

```python
class BankAccount:
    def __init__(self, account_number, holder_name, balance=0):
        self.account_number = account_number
        self.holder_name = holder_name
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited {amount}. New balance: {self.balance}")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if amount > 0 and amount <= self.balance:
            self.balance -= amount
            print(f"Withdrew {amount}. New balance: {self.balance}")
        else:
            print("Invalid withdrawal amount.")

    def inquiry_balance(self):
        print(f"Account balance: {self.balance}")
        return self.balance
```

```python
class SavingsAccount(BankAccount):
    def __init__(self, account_number, holder_name, balance=0, interest_rate=0.02):
        super().__init__(account_number, holder_name, balance)
        self.interest_rate = interest_rate

    def add_interest(self):
        interest = self.balance * self.interest_rate
        self.balance += interest
        print(f"Interest added: {interest}. New balance: {self.balance}")

class CheckingAccount(BankAccount):
    def __init__(self, account_number, holder_name, balance=0, overdraft_limit=500):
        super().__init__(account_number, holder_name, balance)
        self.overdraft_limit = overdraft_limit

    def withdraw(self, amount):
        if amount > 0 and self.balance + self.overdraft_limit >= amount:
            self.balance -= amount
            print(f"Withdrew {amount}. New balance: {self.balance}")
        else:
            print("Withdrawal amount exceeds the overdraft limit.")
```

```python
class MenuItem:
    _id_counter = 1  # Class variable to keep track of unique ID

    def __init__(self, name, description, price, category):
        self.__item_id = MenuItem._id_counter  # Unique ID, encapsulated (hidden)
        MenuItem._id_counter += 1
        self.name = name
        self.description = description
        self.price = price
        self.category = category

    def get_item_id(self):
        return self.__item_id

    def update_item(self, name=None, description=None, price=None, category=None):
        if name:
            self.name = name
        if description:
            self.description = description
        if price:
            self.price = price
        if category:
            self.category = category
        print(f"Item {self.__item_id} updated successfully.")

    def __str__(self):
        return f"{self.name} ({self.category}): ${self.price} - {self.description}"


class Menu:
    def __init__(self):
        self.items = []

    def add_item(self, item):
        self.items.append(item)
        print(f"Added {item.name} to the menu.")

    def update_item(self, item_id, **kwargs):
        for item in self.items:
            if item.get_item_id() == item_id:
                item.update_item(**kwargs)
                return
        print("Item not found.")

    def remove_item(self, item_id):
        for item in self.items:
            if item.get_item_id() == item_id:
                self.items.remove(item)
                print(f"Removed item {item_id} from the menu.")
                return
        print("Item not found.")

    def display_menu(self):
        if not self.items:
            print("The menu is empty.")
        for item in self.items:
            print(item)


class FoodItem(MenuItem):
    def __init__(self, name, description, price, category, cooking_time):
        super().__init__(name, description, price, category)
        self.cooking_time = cooking_time  # Specific attribute for food items

    def __str__(self):
        return f"{self.name} ({self.category}): ${self.price} - {self.description} [Cooking time: {self.cooking_time} minutes]"

class BeverageItem(MenuItem):
    def __init__(self, name, description, price, category, serving_temperature):
        super().__init__(name, description, price, category)
        self.serving_temperature = serving_temperature  # Specific attribute for beverages

    def __str__(self):
        return f"{self.name} ({self.category}): ${self.price} - {self.description} [Serving temperature: {self.serving_temperature}°C]"


# Create menu
menu = Menu()

# Add food items
pasta = FoodItem("Pasta", "Creamy Alfredo pasta", 12.99, "Main Course", cooking_time=15)
burger = FoodItem("Burger", "Juicy beef burger", 9.99, "Main Course", cooking_time=10)

menu.add_item(pasta)
menu.add_item(burger)

# Add beverage items
coffee = BeverageItem("Coffee", "Hot brewed coffee", 3.99, "Beverage", serving_temperature=70)
juice = BeverageItem("Juice", "Fresh orange juice", 4.99, "Beverage", serving_temperature=5)

menu.add_item(coffee)
menu.add_item(juice)

# Display menu
print("\nCurrent Menu:")
menu.display_menu()

# Update an item
menu.update_item(pasta.get_item_id(), price=13.99)

# Remove an item
menu.remove_item(burger.get_item_id())

# Display menu again
print("\nUpdated Menu:")
menu.display_menu()
```

```python
class Room:
    _id_counter = 1  # Class variable to keep track of unique ID

    def __init__(self, room_number, room_type, rate, availability=True):
        self.__room_id = Room._id_counter  # Unique ID, encapsulated (hidden)
        Room._id_counter += 1
        self.room_number = room_number
        self.room_type = room_type
        self.rate = rate
        self.__availability = availability  # Encapsulated availability

    def get_room_id(self):
        return self.__room_id

    def is_available(self):
        return self.__availability

    def book_room(self):
        if self.__availability:
            self.__availability = False
            print(f"Room {self.room_number} has been booked.")
        else:
            print(f"Room {self.room_number} is not available.")

    def check_in(self):
        if not self.__availability:
            print(f"Guest checked into room {self.room_number}.")
        else:
            print(f"Room {self.room_number} is not yet booked.")

    def check_out(self):
        if not self.__availability:
            self.__availability = True
            print(f"Guest checked out of room {self.room_number}. Room is now available.")
        else:
            print(f"Room {self.room_number} is already available.")

    def __str__(self):
        availability = "Available" if self.__availability else "Not Available"
        return f"Room {self.room_number} ({self.room_type}): ${self.rate}/night - {availability}"


class SuiteRoom(Room):
    def __init__(self, room_number, rate, amenities, availability=True):
        super().__init__(room_number, "Suite", rate, availability)
        self.amenities = amenities  # Specific attribute for suite rooms

    def __str__(self):
        amenities_str = ", ".join(self.amenities)
        base_info = super().__str__()
        return f"{base_info} [Amenities: {amenities_str}]"

class StandardRoom(Room):
    def __init__(self, room_number, rate, bed_type, availability=True):
        super().__init__(room_number, "Standard", rate, availability)
        self.bed_type = bed_type  # Specific attribute for standard rooms

    def __str__(self):
        base_info = super().__str__()
        return f"{base_info} [Bed Type: {self.bed_type}]"

# Create hotel rooms
suite_101 = SuiteRoom(101, 250, amenities=["King Bed", "Jacuzzi", "Mini Bar"])
standard_102 = StandardRoom(102, 120, bed_type="Queen Bed")
standard_103 = StandardRoom(103, 100, bed_type="Twin Bed")

# Display room details
print(suite_101)
print(standard_102)
print(standard_103)

# Book a room
suite_101.book_room()

# Try to book the same room again
suite_101.book_room()

# Check in the guest
suite_101.check_in()

# Check out the guest
suite_101.check_out()

# Try to check out again
suite_101.check_out()

# Book a standard room and check-in
standard_102.book_room()
standard_102.check_in()

```

```python
class Flight:
    _id_counter = 1  # Class variable for unique ID

    def __init__(self, flight_number, departure_airport, arrival_airport, departure_time, arrival_time, available_seats):
        self.__flight_id = Flight._id_counter  # Encapsulated unique ID
        Flight._id_counter += 1
        self.flight_number = flight_number
        self.departure_airport = departure_airport
        self.arrival_airport = arrival_airport
        self.departure_time = departure_time
        self.arrival_time = arrival_time
        self.__available_seats = available_seats  # Encapsulated available seats

    def get_flight_id(self):
        return self.__flight_id

    def book_seat(self):
        if self.__available_seats > 0:
            self.__available_seats -= 1
            print(f"Seat booked on flight {self.flight_number}.")
        else:
            print(f"No seats available on flight {self.flight_number}.")

    def cancel_reservation(self):
        self.__available_seats += 1
        print(f"Reservation cancelled on flight {self.flight_number}.")

    def get_remaining_seats(self):
        return self.__available_seats

    def __str__(self):
        return f"Flight {self.flight_number}: {self.departure_airport} -> {self.arrival_airport}, Departure: {self.departure_time}, Arrival: {self.arrival_time}, Seats Available: {self.__available_seats}"

class DomesticFlight(Flight):
    def __init__(self, flight_number, departure_airport, arrival_airport, departure_time, arrival_time, available_seats, baggage_allowance=20):
        super().__init__(flight_number, departure_airport, arrival_airport, departure_time, arrival_time, available_seats)
        self.baggage_allowance = baggage_allowance  # Specific attribute for domestic flights

    def __str__(self):
        base_info = super().__str__()
        return f"{base_info}, Baggage Allowance: {self.baggage_allowance}kg"

class InternationalFlight(Flight):
    def __init__(self, flight_number, departure_airport, arrival_airport, departure_time, arrival_time, available_seats, visa_required=True):
        super().__init__(flight_number, departure_airport, arrival_airport, departure_time, arrival_time, available_seats)
        self.visa_required = visa_required  # Specific attribute for international flights

    def __str__(self):
        base_info = super().__str__()
        return f"{base_info}, Visa Required: {self.visa_required}"

# Create flights
domestic_flight = DomesticFlight("DF123", "JFK", "LAX", "2024-09-01 08:00 AM", "2024-09-01 11:00 AM", 150)
international_flight = InternationalFlight("IF456", "JFK", "LHR", "2024-09-01 06:00 PM", "2024-09-02 06:00 AM", 200, visa_required=True)

# Book seats
domestic_flight.book_seat()
international_flight.book_seat()

# Display flight details
print(domestic_flight)

print(domestic_flight)
print(international_flight)

# Cancel a reservation
domestic_flight.cancel_reservation()

# Display remaining seats
print(f"Remaining seats on {domestic_flight.flight_number}: {domestic_flight.get_remaining_seats()}")
print(f"Remaining seats on {international_flight.flight_number}: {international_flight.get_remaining_seats()}")
```

19. Create a Python program that reads a text file named "paragraph.txt" and counts the occurrences of
each word in the paragraph, displaying the results in alphabetical order.

```python
from collections import Counter
import re

def count_word_occurrences(file_path):
    """Counts the occurrences of each word in the specified file and returns a sorted dictionary."""
    try:
        with open(file_path, 'r') as file:
            text = file.read()

            # Convert text to lowercase to ensure case-insensitive counting
            text = text.lower()

            # Use regular expression to find words (handles punctuation and whitespace)
            words = re.findall(r'\b\w+\b', text)

            # Count occurrences of each word
            word_count = Counter(words)

            # Sort the dictionary by word (alphabetically)
            sorted_word_count = dict(sorted(word_count.items()))

            return sorted_word_count

    except FileNotFoundError:
        print(f"Error: The file {file_path} was not found.")
        return None

    except Exception as e:
        print(f"An error occurred: {e}")
        return None

# Define the file path
file_path = "paragraph.txt"

# Count words and display the results
word_counts = count_word_occurrences(file_path)
if word_counts is not None:
    print("Word occurrences (alphabetically sorted):")
    for word, count in word_counts.items():
        print(f"{word}: {count}")
```
